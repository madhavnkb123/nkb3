```
# Python example to show that base
# class members can be accessed in
# derived class using base class name
class Base(object):
    # Constructor
    def __init__(self, x):
        self.x = x    
class Derived(Base):
    # Constructor
    def __init__(self, x, y):
        Base.x = x 
        self.y = y
    def printXY(self):
       # print(self.x, self.y) will also work
       print(Base.x, self.y)
# Driver Code
d = Derived(10, 20)
d.printXY()
```

```
# Python example to show working of multiple 
# inheritance
class Base1(object):
    def __init__(self):
        self.str1 = "Geek1"
        print ("Base1")
class Base2(object):
    def __init__(self):
        self.str2 = "Geek2"       
        print ("Base2")
class Derived(Base1, Base2):
    def __init__(self):  
        # Calling constructors of Base1
        # and Base2 classes
        Base1.__init__(self)
        Base2.__init__(self)
        print ("Derived")      
    def printStrs(self):
        print(self.str1, self.str2)
ob = Derived()
ob.printStrs()#!/usr/bin/python
```
